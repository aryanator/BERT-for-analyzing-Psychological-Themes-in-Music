```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Load the dataset
dataset = load_dataset("chloeliu/lyrics")

# Convert the dataset to a DataFrame
df = pd.DataFrame(dataset['train'])

# Extract the lyrics column
lyrics = df['lyrics']

# Initialize the CountVectorizer
vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2, max_features=1000)

# Fit and transform the lyrics data
X = vectorizer.fit_transform(lyrics)

# Initialize the LDA model
lda = LatentDirichletAllocation(n_components=5, random_state=42)

# Fit the LDA model to the data
lda.fit(X)

# Display the topics and the top words for each topic
feature_names = vectorizer.get_feature_names_out()
for idx, topic in enumerate(lda.components_):
    print(f"Topic {idx+1}:")
    print([feature_names[i] for i in topic.argsort()[-10:]])
```
