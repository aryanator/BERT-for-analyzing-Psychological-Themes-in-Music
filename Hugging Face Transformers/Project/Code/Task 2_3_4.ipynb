BASIC LDA RECOMMENDATION SYSTEM

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
dataset = load_dataset("chloeliu/lyrics")

# Convert the dataset to a DataFrame
df = pd.DataFrame(dataset['train'][:5000])

# Extract the lyrics column
lyrics = df['lyrics']

# Initialize the CountVectorizer
vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2, max_features=1000)

# Fit and transform the lyrics data
X = vectorizer.fit_transform(lyrics)

# Initialize the LDA model
lda = LatentDirichletAllocation(n_components=10, random_state=42)

# Fit the LDA model to the data
lda.fit(X)

# Transform the data to topic space
topic_distribution = lda.transform(X)

# Calculate cosine similarity between songs
cosine_sim = cosine_similarity(topic_distribution)

# Define a function to recommend songs
def recommend_song(song_index, num_recommendations=5):
    sim_scores = list(enumerate(cosine_sim[song_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations+1]
    song_indices = [i[0] for i in sim_scores]
    return df['track_name'].iloc[song_indices]

# Example: Recommend songs similar to the first song in the dataset
recommendations = recommend_song(0)
print(recommendations)
```

INCLUSION OF FACTORS

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Load the dataset
dataset = load_dataset("chloeliu/lyrics")

# Convert the dataset to a DataFrame and select the first 5000 samples
df = pd.DataFrame(dataset['train'][:5000])

# Extract the lyrics column
lyrics = df['lyrics']

# Initialize the CountVectorizer
vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2, max_features=1000)

# Fit and transform the lyrics data
X = vectorizer.fit_transform(lyrics)

# Initialize the LDA model
lda = LatentDirichletAllocation(n_components=10, random_state=42)

# Fit the LDA model to the data
lda.fit(X)

# Transform the data to topic space
topic_distribution = lda.transform(X)

# Select numeric features
numeric_features = ['len', 'dating', 'violence', 'world/life', 'night/time', 'shake the audience', 'family/gospel', 'romantic', 'communication', 'obscene', 'music', 'movement/places', 'light/visual perceptions', 'family/spiritual', 'like/girls', 'sadness', 'feelings', 'danceability', 'loudness', 'acousticness', 'instrumentalness', 'valence', 'energy', 'age']

# Normalize the numeric features to have a similar scale to the LDA topic distribution
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df[numeric_features])

# Concatenate the LDA topic distribution and the scaled numeric features
combined_features = np.concatenate([topic_distribution, scaled_features], axis=1)

# Calculate cosine similarity between songs using the combined features
cosine_sim_combined = cosine_similarity(combined_features)

# Define a function to recommend songs
def recommend_song_combined(song_index, num_recommendations=5):
    sim_scores = list(enumerate(cosine_sim_combined[song_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations+1]
    song_indices = [i[0] for i in sim_scores]
    return df['track_name'].iloc[song_indices]

# Example: Recommend songs similar to the first song in the dataset
recommendations = recommend_song_combined(0)
print(recommendations)
```

```python
# Example: Recommend songs similar to the first song in the dataset
song = 77
recommendations = recommend_song_combined(song)
print(df['track_name'][song])
print(recommendations)
```

```python
from scipy import stats

# Define a function to calculate average similarity for a song
def avg_similarity(song_index, cosine_sim):
    sim_scores = list(enumerate(cosine_sim[song_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Exclude the song itself
    sim_scores = sim_scores[1:]
    avg_sim = np.mean([score[1] for score in sim_scores])
    return avg_sim

# Calculate average similarities for all songs
avg_similarities_old = [avg_similarity(i, cosine_sim) for i in range(len(df))]
avg_similarities_new = [avg_similarity(i, cosine_sim_combined) for i in range(len(df))]

# Perform paired t-test
t_statistic, p_value = stats.ttest_rel(avg_similarities_old, avg_similarities_new)
alpha = 0.05

if p_value < alpha:
    print("The new system provides better recommendations on average.")
else:
    print("The old system provides better recommendations on average.")
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
```

```python
# Calculate average cosine similarities
avg_sim_old = np.mean(avg_similarities_old)
avg_sim_new = np.mean(avg_similarities_new)

print(f"Average cosine similarity (old system): {avg_sim_old}")
print(f"Average cosine similarity (new system): {avg_sim_new}")

if avg_sim_new > avg_sim_old:
    print("The new system provides better recommendations on average.")
else:
    print("The old system provides better recommendations on average.")
```
